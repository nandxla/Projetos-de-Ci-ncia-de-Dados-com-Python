```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df_anaemic_prediction = pd.read_csv("output.csv")
df_anaemic_prediction.head(10)
```

```python
df_anaemic_prediction["Sex"] = df_anaemic_prediction["Sex"].str.strip() 
```

```python
ohe_sex = pd.get_dummies(df_anaemic_prediction["Sex"]) * 1
df_anaemic_prediction = pd.concat([df_anaemic_prediction, ohe_sex], axis=1)
df_anaemic_prediction.drop(columns=[
    "Sex",
    "Number",
], inplace=True)
```

```python
df_anaemic_prediction.head()
```

```python
anaemic_map = {
    "Yes": 1,
    "No": 0,
}

df_anaemic_prediction["Anaemic"] = df_anaemic_prediction["Anaemic"].map(anaemic_map)
```

```python
y = df_anaemic_prediction["Anaemic"]
X = df_anaemic_prediction.drop(columns=["Anaemic"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
logistic_clf = LogisticRegression()
logistic_clf
```

```python
logistic_clf.fit(X_train, y_train)
y_pred = logistic_clf.predict(X_test)
```
