```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df_credit_risk = pd.read_excel("default of credit card clients.xls")
df_credit_risk.head()
```

```python
df_credit_risk.columns = df_credit_risk.iloc[0]
df_credit_risk.drop(index=0, inplace=True)
```

```python
df_credit_risk.info()
```

```python
df_credit_risk["PAY_0"] = df_credit_risk["PAY_0"].astype('int') 
df_credit_risk["PAY_2"] = df_credit_risk["PAY_2"].astype('int') 
df_credit_risk["PAY_3"] = df_credit_risk["PAY_3"].astype('int') 
df_credit_risk["PAY_4"] = df_credit_risk["PAY_4"].astype('int') 
df_credit_risk["PAY_5"] = df_credit_risk["PAY_5"].astype('int') 
df_credit_risk["PAY_6"] = df_credit_risk["PAY_6"].astype('int') 
```

```python
df_credit_risk["LIMIT_BAL"] = df_credit_risk["LIMIT_BAL"].astype("float")
df_credit_risk[["SEX", "EDUCATION", "MARRIAGE", "AGE"]] = df_credit_risk[["SEX", "EDUCATION", "MARRIAGE", "AGE"]].astype("int")
```

```python
df_credit_risk.head()
```

```python
mpl.rcParams['figure.dpi'] = 100
df_credit_risk[["LIMIT_BAL"]].hist()
plt.ticklabel_format(style='plain', axis='x')

```

```python
df_credit_risk.info()
```

```python
df_credit_risk["EDUCATION"].unique()
```

```python
# df_credit_risk.loc[df_credit_risk["EDUCATION"].isin(0,5,6), "EDUCATION"] = 4
df_credit_risk.replace(to_replace=[0, 5, 6], value=4, inplace=True)
```

```python
education_map = {
    1: "PÃ³s-GraduaÃ§Ã£o",
    2: "GraduaÃ§Ã£o",
    3: "Ensino MÃ©dio",
    4: "Outros"
}
df_credit_risk["EDUCATION_CAT"] = df_credit_risk["EDUCATION"].map(education_map)
```

```python
df_credit_risk.head()
```

```python
edu_ohe = pd.get_dummies(df_credit_risk["EDUCATION_CAT"]).astype('int')
edu_ohe.head()
```

```python
df_with_oh = pd.concat([df_credit_risk, edu_ohe], axis=1)
df_with_oh[["EDUCATION_CAT", "Ensino MÃ©dio", "GraduaÃ§Ã£o", "PÃ³s-GraduaÃ§Ã£o", "Outros"]].head(10)
```

```python
df_with_oh.drop(columns=["EDUCATION_CAT"], inplace=True)
```

```python
df_with_oh.rename(columns={
    "Ensino MÃ©dio": "ENSINO_MEDIO_OH", 
    "GraduaÃ§Ã£o": "GRADUACAO_OH", 
    "PÃ³s-GraduaÃ§Ã£o": "POS_GRADUACAO_OH", 
    "Outros": "OUTROS_OH"
}, inplace=True)
```
